```python
import pandas as pd
import requests
import json
```

```python
request_dict ={"fields": "full_name,epoch,e,a,q,i,om,w,pha,neo,diameter,density",
'sb-class': 'APO,ATE,AMO',
'sb-xfrag': '1'}  # This is the string you need to pass}

response = requests.get("https://ssd-api.jpl.nasa.gov/sbdb_query.api", params = request_dict)
json_resp = response.json()
```

```python
json_resp
```

```python
objects_df = pd.DataFrame.from_dict(json_resp['data'])
objects_df.columns = json_resp['fields']
```
